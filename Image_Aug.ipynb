# Performing some augmentation techniques using imgaug library

```python
import imageio
import imgaug as ia
import imgaug.augmenters as iaa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib
%matplotlib inline
```

```python
path=r'C:\Users\aryas\Downloads\download.jfif'
image = imageio.imread(path)
ia.imshow(image)
```

```python
#drawing bounding boxes 
#drawing bounding boxes benefits in the labelling process for varios computer vision models
from imgaug.augmentables.bbs import BoundingBox, BoundingBoxesOnImage
bounding_box= BoundingBoxesOnImage([BoundingBox(x1=20, x2=170, y1=20, y2=230)], shape=image.shape)
labelled_image=ia.imshow(bounding_box.draw_on_image(image, size=2))
```

```python
#rotating the image without bounding box
rotated=iaa.Affine(rotate=(-20, 50))
rotated_image=rotate.augment_image(image)
ia.imshow(rotated_image)
```

```python
#adding gaussian noise to the image
gaussian_noise=iaa.AdditiveGaussianNoise(20,20)
noise_image=gaussian_noise.augment_image(image)
ia.imshow(noise_image)
```

```python
#cropping the image by a specific percentage
cropped = iaa.Crop(percent=(0.1, 0.35)) 
cropped_image=cropped.augment_image(image)
ia.imshow(cropped_image)
```

```python
#creating a contrast by varying the brightness level
image = imageio.imread(path)
brightness=iaa.GammaContrast(gamma=3.0)
contrast_image =brightness.augment_image(image)
ia.imshow(contrast_image)
```

```python
#shearing the image by a specific amount
sheared = iaa.Affine(shear=(0,40))
shear_image=sheared.augment_image(image)
ia.imshow(shear_image)
```

```python
#flipping the image in horizontal direction
fliph=iaa.Fliplr(p=1.0)
fliph_image= fliph.augment_image(image)
ia.imshow(fliph_image)
```

```python
# flipping image upside down
flipv=iaa.Flipud(p=1.0)
flipv_image= flipv.augment_image(image)
ia.imshow(flipv_image)
```
